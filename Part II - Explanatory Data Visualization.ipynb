# Part II — Explanatory Data Visualization

This notebook contains the explanatory portion of the Prosper Loan Dataset project. It presents polished visualizations and concise narrative that communicate the most important findings from the exploratory analysis.

The explanatory section includes a dataset overview, four presentation-ready visualizations, and a set of conclusions that summarize the key takeaways.

```
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_theme(style='whitegrid')

plt.rcParams['figure.figsize'] = (10, 6)
```

## Investigation Overview

In this explanatory section, we highlight the strongest financial insights from the Prosper Loan dataset. The visualizations focus on loan amounts, borrower interest rates, credit score effects, and the combined impact of income and credit score.

```
# Load the Prosper Loan dataset
loan_df = pd.read_csv('prosperLoanData.csv')

# Select the variables required for the project
selected_columns = [
    'LoanOriginalAmount',
    'BorrowerRate',
    'LoanStatus',
    'IncomeRange',
    'EmploymentStatus',
    'CreditScoreRangeUpper',
    'DebtToIncomeRatio'
]
loan_df = loan_df[selected_columns].copy()

# Convert borrower rate into a percentage for easier interpretation
loan_df['BorrowerRatePercent'] = loan_df['BorrowerRate'] * 100

# Define an ordered income category and clean invalid values
income_order = [
    '$1-24,999',
    '$25,000-49,999',
    '$50,000-74,999',
    '$75,000-99,999',
    '$100,000+'
]
loan_df['IncomeRange'] = pd.Categorical(
    loan_df['IncomeRange'],
    categories=income_order,
    ordered=True
)

loan_df = loan_df.dropna(subset=[
    'LoanOriginalAmount',
    'BorrowerRate',
    'LoanStatus',
    'IncomeRange',
    'CreditScoreRangeUpper',
    'DebtToIncomeRatio'
])
loan_df = loan_df[~loan_df['IncomeRange'].isin(['Not displayed', 'Not employed', '$0'])]

loan_df.reset_index(drop=True, inplace=True)
```

## Dataset Overview

The dataset has been cleaned to include the key fields needed for explanation. Below are summary statistics for the main numerical variables and the selected income and loan status categories.

```
print('Dataset shape:', loan_df.shape)
print('
Numerical summary:')
print(loan_df[['LoanOriginalAmount', 'BorrowerRatePercent', 'CreditScoreRangeUpper', 'DebtToIncomeRatio']].describe())
print('
Income range counts:')
print(loan_df['IncomeRange'].value_counts())
print('
Loan status counts:')
print(loan_df['LoanStatus'].value_counts())
```

## Visualization 1: Distribution of Loan Amounts

**Question:** What overall loan amount pattern should a Prosper stakeholder understand?

**Visualization:** A polished histogram with a mean reference line highlights the primary loan amount distribution.

**Observation:** Most loans are issued in moderate amounts, while very large loans are much less common.

```
plt.figure(figsize=(10, 6))
plt.hist(loan_df['LoanOriginalAmount'], bins=30, color='#5a9', edgecolor='black')
plt.axvline(loan_df['LoanOriginalAmount'].mean(), color='red', linestyle='--', label='Mean Loan Amount')
plt.title('Distribution of Loan Original Amounts', fontsize=16, weight='bold')
plt.xlabel('Loan Original Amount ($)')
plt.ylabel('Number of Loans')
plt.legend()
plt.tight_layout()
plt.show()
```

## Visualization 2: Income vs Loan Amount

**Question:** How does borrower income relate to the size of loans obtained?

**Visualization:** A box plot emphasizes median loan amounts and variability across income ranges.

**Observation:** Loan size rises with income range, showing that higher-income borrowers generally receive larger loans.

```
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=loan_df,
    x='IncomeRange',
    y='LoanOriginalAmount',
    order=loan_df['IncomeRange'].cat.categories,
    palette='coolwarm'
)
plt.title('Loan Amount by Income Range', fontsize=16, weight='bold')
plt.xlabel('Income Range')
plt.ylabel('Loan Original Amount ($)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## Visualization 3: Credit Score vs Borrower Rate

**Question:** What does the relationship between credit score and borrower rate reveal about pricing?

**Visualization:** A regression scatter plot with a trend line highlights the credit score pricing effect.

**Observation:** Higher credit scores are clearly associated with lower borrower rates, showing that creditworthiness is a key determinant of loan cost.

```
plt.figure(figsize=(10, 6))
sns.regplot(
    data=loan_df.sample(n=6000, random_state=42),
    x='CreditScoreRangeUpper',
    y='BorrowerRatePercent',
    scatter_kws={'alpha': 0.25},
    line_kws={'color': 'darkblue', 'linewidth': 2}
)
plt.title('Credit Score vs Borrower Interest Rate', fontsize=16, weight='bold')
plt.xlabel('Credit Score Range Upper')
plt.ylabel('Borrower Rate (%)')
plt.tight_layout()
plt.show()
```

## Visualization 4: Credit Score and Income vs Borrower Rate

**Question:** How do income and credit score together explain borrower rate differences?

**Visualization:** A scatter plot colored by income range shows how the combined factors relate to borrower rate.

**Observation:** High credit score borrowers pay lower rates across income groups, while higher-income borrowers are more likely to appear in lower-rate segments.

```
plt.figure(figsize=(11, 7))
sns.scatterplot(
    data=loan_df.sample(n=8000, random_state=42),
    x='CreditScoreRangeUpper',
    y='BorrowerRatePercent',
    hue='IncomeRange',
    alpha=0.4,
    palette='viridis'
)
plt.title('Credit Score and Income vs Borrower Rate', fontsize=16, weight='bold')
plt.xlabel('Credit Score Range Upper')
plt.ylabel('Borrower Rate (%)')
plt.legend(title='Income Range', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Conclusions

The Prosper Loan dataset analysis shows four key findings: loan amounts are concentrated in moderate ranges, higher-income borrowers tend to take larger loans, credit score and borrower rate are negatively correlated, and income supports but does not replace credit score as a predictor of borrower rate. These insights can guide stakeholders seeking to understand loan pricing and borrower characteristics.
